### 평균 가중 앙상블

```python
import os
import numpy as np
import pandas as pd
from scipy import stats
import plotly.express as px
```

```python
target = 'prediction'
competition = '../input/amex-default-prediction'
submission = pd.read_csv('../input/amex-default-prediction/sample_submission.csv')
```

```python
preds = []
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from scipy import stats
```

```python
def model_weight(model_loss, worst_loss, best_loss):
    return 1- ((best_loss - model_loss)/(best_loss-(worst_loss - 0.01)))
```

```python
best = 0.787
worst = 0.764
w1 = model_weight(0.764, worst, best)
w2 = model_weight(0.787, worst, best) 
w3 = model_weight(0.784, worst, best)

df = pd.read_csv('../input/amex-base-model/20220812_amex_xgboost_submission(0.764).csv')
df['prediction'] = df['prediction']*w1
preds.append((df['prediction']))
print(np.column_stack(preds).shape)

df2 = pd.read_csv('../input/amex-base-model/20220816_amex_lightbgm_submission(0.787).csv')
df2['prediction'] = df2['prediction']*w2
preds.append((df2['prediction']))
print(np.column_stack(preds).shape)

df3 = pd.read_csv('../input/amex-base-model/20220820_amex_catboost_submission(0.784).csv')
df3['prediction'] = df3['prediction']*w3
preds.append((df3['prediction']))
print(np.column_stack(preds).shape)
```

```python
preds
```

```python
%%time
blend_ss = submission.copy()
```

```python
blend_ss['prediction'] = (np.sum(np.column_stack(preds), axis=1) / (w1+w2+w3))
blend_ss
```

```python
blend_ss.to_csv('submission.csv', index=False)
```
